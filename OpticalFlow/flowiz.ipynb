```python
import flowiz as fz
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
files = glob.glob('../3rdparties/flowiz/demo/flo/*.flo')
files
```

```python
img = fz.convert_from_file(files[0])
plt.imshow(img)
```

```python
floArray = fz.read_flow(path=files[0])
print(floArray.shape)
print(floArray.dtype)
```

```python
fl = fz.convert_from_flow(floArray)
plt.imshow(fl)
```

```python
uv = fz.convert_from_flow(floArray, mode='UV')
fig, axarr = plt.subplots(1,2)
axarr[0].imshow(uv[...,0], cmap=plt.get_cmap('binary'))
axarr[1].imshow(uv[...,1], cmap=plt.get_cmap('binary'))
```
